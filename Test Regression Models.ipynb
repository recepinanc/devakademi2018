```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_json('dev_akademi.json')
```

```python
dataset.head()
```

```python
dataset = dataset[['ad_bid_price_kurus','ad_call_to_action', 'ad_daily_budget_kurus', 'ad_title', 'ad_id', 'ad_description', 'event_type', 'event_category']]
```

```python
dataset.info()
```

```python
dataset.dropna().info()
```

```python
df = dataset.dropna()
```

```python
df = df[df.ad_call_to_action != ""]
df = df[df.event_type != ""]
```

```python
modify_df = df[['ad_bid_price_kurus', 'ad_daily_budget_kurus', 'ad_id', 'event_type']]
```

```python
modify_df = pd.concat([modify_df, pd.get_dummies(modify_df.event_type, prefix='event')], axis=1)
```

```python
modify_df.info()
```

```python
modify_df
```

```python
grouped_df = modify_df.groupby(['ad_bid_price_kurus','ad_daily_budget_kurus'])['event_CLICK', 'event_IMPRESSION'].sum()
```

```python
grouped_df
```

```python
grouped_df = grouped_df.assign(conversion_rate= (grouped_df.event_CLICK/grouped_df.event_IMPRESSION)*100)
```

```python
ready_df = grouped_df.reset_index().drop(['event_CLICK','event_IMPRESSION'], axis=1)
```

```python
ready_df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = ready_df.drop(['conversion_rate'], axis=1)
y = ready_df.conversion_rate
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor()
```

```python
y_train
```

```python
regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
y_pred = regressor.predict(X_test)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```
